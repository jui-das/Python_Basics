```python
#Shopping Bill

Soap = float(input("Enter the price of soap :"))
Shampoo = float(input("Enter the price of shampoo :"))
Toothpaste = float(input("Enter the price of toothpaste :"))
quantity_soap = float(input("Enter the quantity of soap :"))
quantity_shampoo = float(input("Enter the quantity of shampoo :"))
quantity_toothpaste = float(input("Enter the quantity of toothpaste :"))
total_bill = Soap * quantity_soap + Shampoo * quantity_shampoo + Toothpaste * quantity_toothpaste
discount = total_bill * 15/100
final_bill = total_bill - discount
print("Total bill :", total_bill)
print("Discount :", discount)
print("You have to pay :", final_bill)
```
